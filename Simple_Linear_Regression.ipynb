```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Correct URL for the regression dataset (cgpa vs package)
url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day48-simple-linear-regression/placement.csv"

try:
    df = pd.read_csv(url)
    print("Regression Dataset loaded successfully!")
    display(df.head())
except Exception as e:
    print(f"Error loading dataset: {e}")
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in lpa)')
plt.title('CGPA vs Package')
# plt.grid(True)
plt.show()
```

```python
x = df.iloc[:,:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[5].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(x_train, lr.predict(x_train), color='red', label='Regression Line')
plt.legend()
plt.xlabel('CGPA')
plt.ylabel('Package (in lpa)')
plt.title('CGPA vs Package')
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```

```python
# y = mx + b

m*8.35 + b
```
