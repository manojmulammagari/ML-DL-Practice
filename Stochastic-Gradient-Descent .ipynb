```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
class SGDRegressor:
    def __init__(self, learning_rate=0.01, epochs=100):
        # Initializing the 'brain' of our model with a learning speed and practice count
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, x_train, y_train):
        # 1. We start with a baseline guess: intercept at 0 and all weights at 1
        self.intercept_ = 0
        self.coef_ = np.ones(x_train.shape[1])

        # 2. Outer loop: This is like the number of school years (epochs)
        for i in range(self.epochs):
            # 3. Inner loop: Looking at every single row of data
            for j in range(x_train.shape[0]):
                # 4. Pick one random person (sample) to learn from right now
                idx = np.random.randint(0, x_train.shape[0])

                # 5. Make a prediction using our current 'guess' formula
                y_hat = np.dot(x_train[idx], self.coef_) + self.intercept_

                # 6. Check the error: How far off were we from the real answer?
                # We calculate 'derivatives' to see which direction we need to move
                intercept_der = -2 * (y_train[idx] - y_hat)
                coef_der = -2 * (y_train[idx] - y_hat) * x_train[idx]

                # 7. The Update: Slightly nudge the weights based on the error
                # This is the actual 'learning' part of AI
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                self.coef_ = self.coef_ - (self.lr * coef_der)

        print(f"Training Finished. The model is now ready!")
        print( self.intercept_,self.coef_)

    def predict(self, x_test):
        # Use the final math we learned to guess outcomes for new data
        return np.dot(x_test, self.coef_) + self.intercept_
```

```python
# 1. Initialize our custom AI model with a 0.005 learning speed
SGD = SGDRegressor(learning_rate=0.01, epochs=40)

# 2. Feed the data to the model so it can learn patterns
SGD.fit(x_train, y_train)
```

```python
# 3. Ask the model to predict outcomes for people it wasn't trained on
y_pred = SGD.predict(x_test)
```

```python
# 4. Final Score: 1.0 is a perfect score.
# Getting ~0.43 means the model successfully learned the core trends!
print(f"R2 Accuracy Score: {r2_score(y_test, y_pred)}")
```
